```python
cords = []

for d in ds_json:
    for gw in d['rxInfo']:
        if gw['gatewayId']=='7076ff005607214b':
            cords.append([gw['location']['latitude'], gw['location']['longitude']])
            print(gw)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame(cords, columns=['lat','lon'])

# Plot
plt.figure(figsize=(10, 6))
plt.scatter(df['lon'], df['lat'], color='red', label='GW locations')
plt.title('Gateway coordinates')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Annotate points with city names
# for i, city in enumerate(df['city']):
    # plt.text(df['longitude'][i], df['latitude'][i], city)

plt.grid()
plt.legend()
plt.show()
```
