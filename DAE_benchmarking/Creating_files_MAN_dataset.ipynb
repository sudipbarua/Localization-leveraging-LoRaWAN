```python
from collections import OrderedDict 
import pandas as pd
from sklearn.model_selection import train_test_split 
import numpy as np

random_state = 42

my_list_of_dict_X =[]
my_list_of_dict_Y =[]
with open('files/M/offline.final.trace.txt') as f:
    lines = f.readlines()
for line in lines:    
    if not line.startswith('#'):
        t = line.split(';')
        my_dict_X = OrderedDict()
        my_dict_Y = OrderedDict()
        for element in t:
            if element.startswith('t'):
                  i=1
            elif element.startswith('id'):
                  i=1
            elif element.startswith('pos'):
                pos = element.split("=")[1]
                my_dict_Y['x'] = float(pos.split(",")[0])
                my_dict_Y['y'] = float(pos.split(",")[1])
            elif element.startswith('degree'):
                  i=1
            else:
                temp = element.split("=")[1]
                my_dict_X[element.split("=")[0]] = int(temp.split(",")[0])
        my_list_of_dict_X.append(my_dict_X)   
        my_list_of_dict_Y.append(my_dict_Y)   
df_X = pd.DataFrame(my_list_of_dict_X)
df_Y = pd.DataFrame(my_list_of_dict_Y)


df_X = pd.DataFrame(my_list_of_dict_X)
df_Y = pd.DataFrame(my_list_of_dict_Y)

x = df_X.fillna(0).to_numpy()
y = df_Y.to_numpy()
print("x size: ", x.shape)
print("y size: ", y.shape)    
y, y_indeces = np.unique(y, return_index=True,axis=0)
print(y_indeces.shape)
x = x[y_indeces]
print('keeping unique')
print("x size: ", x.shape)
print("y size: ", y.shape) 
print() 

x = pd.DataFrame(x)
y = pd.DataFrame(y)


x_train, x_test_val, y_train, y_test_val = train_test_split(x, y, test_size=0.3, random_state=random_state)
x_test, x_val, y_test, y_val = train_test_split(x_test_val, y_test_val, test_size=0.5, random_state=random_state)


minimum = int(x_train.min().min() - 1)
print('minimum')
print(minimum)
x_train = x_train.replace(0,minimum)
x_val = x_val.replace(-100,minimum)
x_test = x_test.replace(-100,minimum)


x_train.to_csv('files/M/x_train.csv', index=False)
x_val.to_csv('files/M/x_val.csv', index=False)
x_test.to_csv('files/M/x_test.csv', index=False)

y_train.to_csv('files/M/y_train.csv', index=False)
y_val.to_csv('files/M/y_val.csv', index=False)
y_test.to_csv('files/M/y_test.csv', index=False)
```
