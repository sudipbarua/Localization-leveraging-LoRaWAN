```python
import pandas as p
import numpy as np
from sklearn.model_selection import train_test_split 

default_NaN = -150
random_state = 42

x_train = p.read_csv('files/DSI/tj_rss.csv',header=None)
y_train = p.read_csv('files/DSI/tj_crd.csv',header=None)   
x_val_test = p.read_csv('files/DSI/rm_rss.csv',header=None)
y_val_test = p.read_csv('files/DSI/rm_crd.csv',header=None)  

x_val_test = x_val_test.values
y_val_test = y_val_test.values
x_train = x_train.values
y_train = y_train.values

print("x_train size: ", x_train.shape)

print("x_val_test size: ", x_val_test.shape)
print("y_val_test size: ", y_val_test.shape)    
y_val_test, y_val_test_indeces = np.unique(y_val_test, return_index=True,axis=0)
print(y_val_test_indeces.shape)
x_val_test = x_val_test[y_val_test_indeces]
print('keeping unique')
print("x_val_test size: ", x_val_test.shape)
print("y_val_test size: ", y_val_test.shape) 
print() 

ratio = 0.5
x_val, x_test, y_val, y_test = train_test_split(x_val_test, y_val_test, test_size=ratio, random_state=random_state)
      
print("x_train size: ", x_train.shape)
print("x_val size: ", x_val.shape)
print("x_test size: ", x_test.shape)


x_train = np.where(x_train==default_NaN, 0, x_train) 
minimum = int(x_train.min() - 1)
print('minimum')
print(minimum)
x_train = np.where(x_train==0, minimum, x_train) 
x_val = np.where(x_val==default_NaN, minimum, x_val) 
x_test = np.where(x_test==default_NaN, minimum, x_test) 


p.DataFrame(x_train, index=None, columns=None).to_csv('files/DSI/x_train.csv', index=False)
p.DataFrame(x_val, index=None, columns=None).to_csv('files/DSI/x_val.csv', index=False)
p.DataFrame(x_test, index=None, columns=None).to_csv('files/DSI/x_test.csv', index=False)

p.DataFrame(y_train, index=None, columns=None).to_csv('files/DSI/y_train.csv', index=False)
p.DataFrame(y_val, index=None, columns=None).to_csv('files/DSI/y_val.csv', index=False)
p.DataFrame(y_test, index=None, columns=None).to_csv('files/DSI/y_test.csv', index=False)

print("done")
```
